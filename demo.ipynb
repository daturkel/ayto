# AYTO quickstart guide

Using the `ayto` library is simple: everything you need is in the `AYTO` class.

```python
from ayto import AYTO
```

To initialize the class, pass it a list of guys and girls. Names must be unique within the list of guys and the list of girls.

```python
guys = [
    "Al",
    "Bob",
    "Chuck",
    "Dale",
    "Evan",
    "Frank",
    "Graham",
    "Hugh",
    "Ike",
    "James",
    "Kirk",
]

girls = [
    "Lauren",
    "Mandy",
    "Nora",
    "Olivia",
    "Pam",
    "Quinn",
    "Riley",
    "Sam",
    "Tara",
    "Uma",
    "Violet",
]
```

```python
season = AYTO(guys, girls)
```

If we want to know how many scenarios are remaining, we can access the `num_scenarios` attribute:

```python
print(f"There are {season.num_scenarios:,} possible scenarios!")
```

If we want to see the probabilities of all couples, we can access the `probabilities` attribute:

```python
season.probabilities
```

All odds are the same! Not very surprising because we haven't provided any information yet.

Let's input the outcome of a Truth Booth where Al and Mindy find out they're not a match.

We use the `apply_truth_booth(guy, girl, match)` method to input this information. The method returns the number of remaining scenarios. 

```python
remaining_scenarios = season.apply_truth_booth("Al", "Mandy", False)

print(f"{remaining_scenarios:,} left")
```

Down from 40 million to 36 million, not bad! Let's look at the probabilities:

```python
season.probabilities
```

Notice the big fat `0.0` in the Al x Mindy cell and how it's impacted Al and Mindy's odds of being with everyone else.

Now let's apply a Matchup Ceremony with the `apply_matchup_ceremony(matchup, beams)` method. We'll apply a ceremony that got 3 beams (meaning three of the pairs were correct).

```python
matchup = [
    ("Al", "Nora"),
    ("Bob", "Mandy"),
    ("Chuck", "Olivia"),
    ("Dale", "Lauren"),
    ("Evan", "Pam"),
    ("Frank", "Quinn"),
    ("Graham", "Riley"),
    ("Hugh", "Quinn"),
    ("Ike", "Tara"),
    ("James", "Uma"),
    ("Kirk", "Violet"),
]
remaining_scenarios = season.apply_matchup_ceremony(matchup, 3)

print(f"{remaining_scenarios:,} left")
```

Wow, that made a huge difference: from 36 million to 2 million scenarios.

Let's see the probabilities:

```python
season.probabilities
```

That's a little hard to parse. With the help of the `seaborn` library (`pip install seaborn` if you don't already have it), we can get a better look at this:

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(8, 7), constrained_layout=True)
g = sns.heatmap(season.probabilities, annot=True, fmt="0.1%", ax=ax, square=True)
```

## Saving and loading

You can save your `AYTO` object with the `save` method, which saves it as a lightweight json file.

```python
season.save("my_season")
```

Then load with `AYTO.load`. Everything is restored.

```python
loaded_season = AYTO.load("my_season")

loaded_season.probabilities
```

## Trying out scenarios

If you've got a hunch about something, you can test out how it would impact the couple probabilities with `try_partial_scenario`.

```python
num_scenarios, probability, couple_probabilities = season.try_partial_scenario(
    matches=[("Bob", "Lauren"), ("Al", "Nora")]
)

print(f"There are {num_scenarios} where Bob and Lauren are a couple and Al and Nora are a couple.")
print(f"That's {round(100 * probability, 1)}% of the current remaining scenarios.")
```

This is what the couple probabilities look like in that case:

```python
couple_probabilities
```

The `try_partial_scenario` method lets you combine hypothetical matches and hypothetical non-matches, or just one or the other.

```python
# try some non-matches
_, _, _ = season.try_partial_scenario(non_matches=[("Al", "Nora"), ("James", "Mandy")])

# try a match and a non-match
_, _, _ = season.try_partial_scenario(
    matches=[("Al", "Nora")], non_matches=[("James", "Mandy")]
)
```

Importantly, your `AYTO` object is not modified by `try_partial_scenario`, so you can test out whatever hypotheticals you want without impacting your running probabilities stored in the `probabilities` attribute.

## Other notes

If you want to apply a bunch of Truth Booths/Matchup Ceremonies without calculating the intermediate probabilities, you can do so by passing `calc_probs=False`, then manually calculate them with the `calculate_probabilities()` method at the end. This will be faster.

```python
tbs = [("Al", "Tara", False), ("Graham", "Uma", False), ("Bob", "Violet", True)]
```

```python
%%time

season = AYTO(guys, girls)

# the slow way, calculating probabilities after every step
for guy, girl, match in tbs:
    season.apply_truth_booth(guy, girl, match)
```

```python
%%time

season = AYTO(guys, girls)

# the fast way, calculating probabilties only at the end
for guy, girl, match in tbs:
    season.apply_truth_booth(guy, girl, match, calc_probs=False)

season.calculate_probabilities()
```
